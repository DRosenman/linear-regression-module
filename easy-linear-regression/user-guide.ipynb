## Creating Dataset Object
### Creating dataset object from a list of x-coordinates and a list of y-coordinates:
If you have n datapoints (x1,y1),(x2,y2),...,(xn,yn)


```python
x = [x1,x2,...,xn]
y = [y1,y2,...,yn]

datasetname = lr.dataset(x,y)
```

*Example:* Data points: (1.0,10.2),(2.0,15.3),(3.0,30.3),(4.0,45.2)

````python
#Data points: (1.0,10.2),(2.0,15.3),(3.0,30.3),(4.0,45.2)
x_coordinates = [1.0,2.0,3.0,4.0]
y_coordinates = [10.2,15.3,30.3,45.2]

dataset1 = lr.dataset(x_coordinates,y_coordinates)
````

### Creating dataset object from a single list of data points:
If you have n datapoints (x1,y1),(x2,y2),...,(xn,yn)


```python
datapoints = [(x1,y1),(x2,y2),...,(xn,yn)]

datasetname = lr.dataset(datapoints)
```

*Example:* Data points: (1.0,10.2),(2.0,15.3),(3.0,30.3),(4.0,45.2)

````python
#Data points: (1.0,10.2),(2.0,15.3),(3.0,30.3),(4.0,45.2)
points = [(1.0,10.2),(2.0,15.3),(3.0,30.3),(4.0,45.2)]
daves_data = lr.dataset(points)
````
